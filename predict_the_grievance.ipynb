```
#reading the data
import pandas as pd
train=pd.read_csv('/content/drive/MyDrive/hackerearth ml/hackerearth ml/dataset/train.csv');
train.head()
```

```
test=pd.read_csv('/content/drive/MyDrive/hackerearth ml/hackerearth ml/dataset/test.csv');
test.head()
```

```
#combining both data frames to apply feature engineering technique
test['importance']=-1
train['data_from']='train'
test['data_from']='test'
total_data=pd.concat([train,test],axis=0)
```

```
total_data.describe()
```

```
for feature in total_data:
  print(feature," null values are =",total_data[feature].isnull().sum())
```

```
for feature in total_data:
  print(feature," unique values are =",total_data[feature].nunique())
```

```
total_data.head()
```

```
#combining issues
combine_issues=['issue.0','issue.1','issue.10','issue.11','issue.12','issue.13','issue.14','issue.15','issue.16','issue.17','issue.18','issue.19','issue.2','issue.20','issue.21','issue.22','issue.23','issue.24','issue.25','issue.26','issue.3','issue.4','issue.5','issue.6','issue.7','issue.8','issue.9']
combine_issues_df=total_data[combine_issues]
#replacing NaN with ''
combine_issues_df.fillna('',inplace=True)
combine_issues_df['issues'] = combine_issues_df[combine_issues].apply(lambda x: '. '.join([data for data in x if data != '']), axis=1)
#removing combine_issues from total_data and combine_issues_df
total_data.drop(combine_issues,axis=1,inplace=True)
combine_issues_df.drop(combine_issues,axis=1,inplace=True)
#concatinate total_data with combine_issues_df
total_data=pd.concat([total_data,combine_issues_df],axis=1)
total_data.head()
```

```
#converting categorical values to lower case
for feature in total_data.columns:
  if total_data[feature].dtype=='object':
    total_data[feature] = total_data[feature].str.lower()
```

```
total_data.head()
```

```
country_map= total_data[['respondentOrderEng','country.name']].set_index('country.name').T.to_dict('list')
```

```
country_map
```

```
 respondent_map = total_data[['respondentOrderEng','respondent.0']].set_index('respondent.0').T.to_dict('list')
```

```
respondent_map
```

```
country_dict={}
for val in (country_map,respondent_map):
  country_dict.update(val)
country_dict
```

```
country_dict = {k: v for k, v in country_dict.items() if pd.notna(k)}
total_data['respondent.0'] = total_data['respondent.0'].apply(lambda x: country_dict[x][0])
total_data['respondent.1'] = total_data['respondent.1'].apply(lambda x: country_dict[x][0] if pd.notnull(x) else x)
total_data['respondent.2'] = total_data['respondent.2'].apply(lambda x: country_dict[x][0] if pd.notnull(x) else x)
total_data['respondent.3'] = total_data['respondent.3'].apply(lambda x: country_dict[x][0] if pd.notnull(x) else x)
total_data['respondent.4'] = total_data['respondent.4'].apply(lambda x: country_dict[x][0] if pd.notnull(x) else x)
del total_data['respondentOrderEng']
```

```
country_dict
```

```
total_data.head()
```

```
total_data['itemid'] = total_data['itemid'].apply(lambda x: x[4:7])
total_data['sharepointid'] = total_data['sharepointid'].apply(lambda x: str(x)[:3])
total_data['total_respondents'] =total_data[['respondent.0','respondent.1','respondent.2','respondent.3','respondent.4']].isna().sum(axis=1)
total_data['total_respondents']=5-total_data['total_respondents']
```

```
total_data.head()
```

```
total_data['daysbetween_intro_decision'] = (pd.to_datetime(total_data['decisiondate']) - pd.to_datetime(total_data['introductiondate'])).dt.days
total_data['daysbetween_intro_judgement'] = (pd.to_datetime(total_data['judgementdate']) - pd.to_datetime(total_data['introductiondate'])).dt.days
total_data['daysbetween_decision_judgement'] = (pd.to_datetime(total_data['judgementdate']) - pd.to_datetime(total_data['decisiondate'])).dt.days
total_data.drop(['decisiondate','introductiondate','judgementdate'], axis=1, inplace=True)
```

```
total_data.head()
```

```
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
total_data['doctypebranch'] = le.fit_transform(total_data['doctypebranch'])
total_data['separateopinion'] = le.fit_transform(total_data['separateopinion'])
total_data['typedescription'] = le.fit_transform(total_data['typedescription'])
```

```
total_data.head()
```

```
for feature in total_data.columns:
  if(total_data[feature].nunique()==1):
    del total_data[feature]
```

```
remove_cols =['parties.0', 'country.alpha2', 'parties.1', 'country.name', 'docname', 'appno', 'ecli', 'kpdate', 'originatingbody_name']
for col in remove_cols:
  if col in total_data.columns:
    total_data.drop(col, axis=1, inplace=True)
```

```
from catboost import CatBoostClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import accuracy_score
```

```
train_data=total_data.query('data_from=="train"')
```

```
train_data=train_data.drop(['data_from','issues'],axis=1)
```

```
train_data.head()
```

```
X_train,X_test,Y_train,Y_test=train_test_split(train_data.drop(['importance'], axis=1),train_data['importance'],test_size=0.2,stratify=train_data['importance'])
```

```
params = {
          'iterations':[250,100,500,1000],
          'learning_rate':[0.03,0.001,0.005,0.01,0.1,0.2,0.3], 
          }
```

```
lgb=CatBoostClassifier()
```

```
grid=GridSearchCV(lgb,param_grid=params,cv=10)
grid.fit(X_train,Y_train)
```

```
grid.best_params_
```

```
lgb=CatBoostClassifier(iterations=1000,learning_rate=0.3)
lgb.fit(X_train,Y_train)
pred=lgb.predict(X_test)
print(classification_report(Y_test,pred))
```

```
tst = total_data.query('data_from == "test"').drop(['data_from','issues', 'importance'] , axis=1)
preds = lgb.predict(tst)
preds
```
